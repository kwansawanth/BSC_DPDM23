###ข้อ1 ช่วงเวลา Christmas and New year (1 Dec - 31 Dec)

```python
import datetime
```

```python
data['InvoiceDate']
```

```python
#data['InvoiceDate']=pd.to_datetime(data.InvoiceDate, format='%m/%d/%Y')
```

```python
data['InvoiceDate']=pd.to_datetime(data.InvoiceDate, format='%m/%d/%Y %H:%M' )
```

```python
data['Date'] = data['InvoiceDate'].dt.date
```

```python
data['Date']=pd.to_datetime(data.Date, format='%Y/%m/%d')
```

```python
data
```

```python
#Select 1-31 Dec
December_1 = data[(data['Date'].dt.month == 12) & (data['Date'].dt.day >= 1) & (data['Date'].dt.day <= 31)]
#ปริ้นเพื่อตรวจสอบ
print(December_1)
```

```python
transactions1 = December_1.groupby('InvoiceNo')['Description'].apply(list).reset_index()
print(transactions1)
```

```python
transactions1_apply = transactions1['Description'].apply(lambda x: [str(item) for item in x])
transactions1_apply
```

```python
frequentItemset = list(apriori(transactions1_apply, min_support=0.036))

for record in frequentItemset:
    if len(record.items) > 1:
        print('yes')
        print(f"Items: {', '.join(record.items)}")
        print(f"Support: {record.support:.4f} ({record.support*100:.2f}%)")

        for i, stat in enumerate(record.ordered_statistics, 1):
            items_base = 'None' if not stat.items_base else ', '.join(stat.items_base)
            items_add = ', '.join(stat.items_add)
            confidence = stat.confidence
            lift = stat.lift

            print(f"Rule {i}: ({items_base} -> {items_add} : Confidence={confidence:.4f}, Lift={lift:.4f})")
        print()
```
